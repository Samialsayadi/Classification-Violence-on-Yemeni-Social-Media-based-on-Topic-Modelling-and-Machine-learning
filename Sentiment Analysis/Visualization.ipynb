```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
file=pd.read_csv('finally_Data.csv')
```

```python
x = []
y = []

for row in file:
    x.append(row[0])
    y.append(row[1])
plt.bar(x, y, color = 'g', width = 0.72, label = "Age")
plt.xlabel('Text')
plt.ylabel('Topic')
plt.title('Ages of different persons')
plt.legend()
plt.show()
```
